```python
from sklearn.preprocessing import OneHotEncoder 


X_cat.drop('CLNT_JOB_POSITION', axis = 1, inplace = True)


ohe = OneHotEncoder(handle_unknown = 'ignore') #sparse_output = False, 
X_cat_ohe = ohe.fit_transform(X_cat).toarray()
```

```python
import matplotlib.pyplot as plt

attr = X_ready.columns.tolist()
attr_df = pd.DataFrame(attr, columns=['attr_name'])
coef_df = pd.DataFrame(np.abs(edu.coef_).T, columns=['attr_meaning'])
rslt_df = pd.concat([attr_df, coef_df], axis = 1)
top10 = rslt_df.sort_values('attr_meaning', ascending=False)[:10]

def f_importances(weights, names, top=-1):
    weights, names = zip(*sorted(list(zip(weights, names))))

    # Show all features
    if top == -1:
        top = len(names)
    plt.figure(figsize=(12, 5))
    plt.barh(range(top), weights[::-1][0:top], align='center')
    plt.yticks(range(top), names[::-1][0:top])
    plt.show()

    
f_importances(top10['attr_meaning'], top10['attr_name'])
```
